## SHAP

```python
df_shap = pd.read_json("../results/feature-importance/metrics-shap.jsonl", lines=True)
df_shap = df_shap.join(pd.json_normalize(df_shap["shap_values"])).drop(columns=["shap_values"])
df_shap.head()
```

```python
dd = df_shap.melt(id_vars=["algorithm"])
dd = dd.pivot_table(index="variable", columns="algorithm")
dd.style.background_gradient()
```

```python
import numpy as np
import pandas as pd
```

```python
df_train = pd.read_json("../scripts/metrics.jsonl", lines=True)
df_train.head()
```

```python
df_test = pd.read_csv("../scripts/results.csv")
df_test.head()
```

```python
df_train["filename"] = df_train["params_dataset"] + "_" + df_train["params_model"] + "_" + df_train["params_loss"] + "_" + df_train["params_depth"].astype(str) + ".keras"
df_train.head()
```

```python
df_train["filename"]
```

```python
df_combined = pd.merge(df_train[["filename", "training_time"]], df_test, on="filename")
df_combined[["params_dataset", "params_model", "params_loss", "params_depth"]] = df_combined["filename"].str.replace("resnet_bias", "resnet-bias").str.split("_", expand=True)
df_combined["params_depth"] = df_combined["params_depth"].str.replace(".keras", "").astype(int)
df_combined.to_excel("summary-rev.xlsx")
df_combined.head()
```

```python
dd = df_combined.pivot_table(index=["params_model", "params_loss", "params_dataset"], columns=["params_depth"], values=["mcc", "f1"])
dd.style.background_gradient()
```
